```python
# создадим новую таблицу, где разъединим на новые строки "склеенные" ообщения, по беглому осмотру через ";", "/"
df_split = data.assign(message=data['message'].str.split('; |/ ')).explode('message').reset_index(drop=True)
df_split.tail(60)
```

```python
# создадим список, где встречается "Краснодар"
pattern = re.compile('Краснодар')
matches = [x for x in df_split['message'] if pattern.match(x)]
```

# Извлечение номера телефона

```python
# регулярное выражение для вырезки номеров телефона в трёх форматах
pattern_phone = '([7-8]\d{10}|[7-8]\[ ]\d{10}|[7-8]\D{,2}\d{3}\D{,2}\d{3}\D\d{2}\D\d{2})'
```

```python
df_split['номер телефона кандидата'] = df_split['message'].str.extract(pattern_phone, expand=True)
df_split['номер телефона кандидата'].isnull().sum()
```

```python
df_split[df_split['номер телефона кандидата'].isnull() == False]['номер телефона кандидата'].value_counts().head(60)
```

Вроде все номера корректны

# Извлечение ФИО

```python
# берём слова, начинающиеся с заглавных букв, 2 или 3 слова через пробел
pattern_name = '([A-Я][а-я|ё]+\ [A-Я][а-я|ё]+\ [A-Я][а-я|ё]+|[A-Я][а-я|ё]+\ [A-Я][а-я|ё]+)'
```

```python
df_split['ФИО'] = df_split['message'].str.extract(pattern_name, expand=True)
```
